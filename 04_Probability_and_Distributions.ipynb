Bayesian rule and total probaiblity theorem:

In a water quality monitoring project, a civil–environmental engineer wants to estimate the probability that a river section is contaminated given a sensor alarm from an IoT water quality station.
Probability that the river is contaminated:
P(C)=0.20. The sensor accuracy is 0.90 and its inaccuracy if 0.05 i.e.,
If the river is not contaminated, the sensor raises an alarm with probability
P(A∣C)=0.05. What is the probability of contamination if the sensor raises alarm?

```python
# Given probabilities
P_C = 0.20          # River contaminated
P_notC = 0.80       # River not contaminated
P_A_given_C = 0.90  # Alarm given contamination
P_A_given_notC = 0.05  # Alarm given no contamination

# 1️⃣ Total Probability Theorem: P(A)
P_A = P_A_given_C * P_C + P_A_given_notC * P_notC

# 2️⃣ Bayes’ Theorem: P(C|A)
P_C_given_A = (P_A_given_C * P_C) / P_A

print(f"Probability of alarm, P(A): {P_A:.3f}")
print(f"Probability river is contaminated given alarm, P(C|A): {P_C_given_A:.3f}")
```

Bernoulli Distribution:

The Bernoulli distribution models binary outcomes (success/failure) with a single trial.

Transportation Engineering: Whether a vehicle at an intersection complies with a speed limit (1 = complies, 0 = exceeds), with probability based on observed compliance rates.
Construction Management: Whether a daily site inspection identifies a safety violation (1 = violation found, 0 = none), reflecting random checks in logistics modeling.
Environmental Engineering: Whether a single air quality sample exceeds pollutant thresholds (1 = exceeds, 0 = within limits), used in monitoring compliance with standards.

Problem:
In transportation engineering, consider the probability that a vehicle arriving at a traffic signal during the red phase stops completely (success) or runs the red light (failure). The probability of stopping is 0.95. Simulate 100 vehicle arrivals and calculate the proportion that stop.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
data = np.random.binomial(1, 0.95, 100)  # 1 for stop, 0 for run
proportion_stop = np.mean(data)
print(f"Generated data (first 10): {data[:10]}")
print(f"Proportion that stop: {proportion_stop:.2f}")

# Plot distribution (PMF as bar plot)
sns.countplot(x=data)
plt.title('Bernoulli Distribution: Stops (1) vs. Runs (0)')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.show()
```

Binomial Distribution:


The Binomial distribution models the number of successes in a fixed number of independent Bernoulli trials.

Transportation Engineering: Number of vehicles in a sample of 20 that pass an emissions test, where each has a fixed probability of passing, applicable in traffic control effectiveness studies.
Construction Management: Number of building inspections in a set of 10 that pass without issues, as in probabilistic models for inspector reliability.
Environmental Engineering: Number of water samples in a batch of 15 that show contamination, used in assessing site pollution risks.

Problem: A highway toll booth processes 20 vehicles in a batch. Each vehicle has a 0.8 probability of having an electronic toll tag (success, quick passage). Model the number of tagged vehicles in a batch and find the probability of exactly 16 tagged vehicles. Simulate 50 batches.

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
n, p = 20, 0.8
data = np.random.binomial(n, p, 50)
prob_exact_16 = binom.pmf(16, n, p)
print(f"Generated data (first 10 batches): {data[:10]}")
print(f"Mean number of tagged vehicles: {np.mean(data):.2f}")
print(f"Probability of exactly 16 tagged: {prob_exact_16:.4f}")

# Plot distribution (histogram with PMF overlay)
sns.histplot(data, bins=range(n+2), stat='probability', kde=False)
x = np.arange(0, n+1)
plt.plot(x, binom.pmf(x, n, p), 'ro-', label='PMF')
plt.title('Binomial Distribution: Tagged Vehicles')
plt.xlabel('Number of Tagged Vehicles')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

Geometric Distribution:

The Geometric distribution models the number of trials until the first success.

Transportation Engineering: Number of consecutive vehicles observed until one violates a traffic rule, such as running a red light, in traffic flow analysis.
Construction Management: Number of contract bids submitted until the first project win, reflecting probabilistic logistics in bidding processes.
Environmental Engineering: Number of soil tests conducted until the first detection of hazardous material, in site remediation planning.

Problem: In bridge inspection, the probability of finding a structural defect in a single inspection is 0.1. Model the number of inspections until the first defect is found. Simulate for 100 bridges and find the expected number of inspections.

```python
import numpy as np
from scipy.stats import geom
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
p = 0.1
data = np.random.geometric(p, 100)  # Trials until first success
expected = 1 / p
print(f"Generated data (first 10): {data[:10]}")
print(f"Mean inspections until defect: {np.mean(data):.2f}")
print(f"Theoretical expected value: {expected:.1f}")

# Plot distribution (histogram with PMF overlay)
sns.histplot(data, bins=range(1, max(data)+2), stat='probability', kde=False)
x = np.arange(1, max(data)+1)
plt.plot(x, geom.pmf(x, p), 'ro-', label='PMF')
plt.title('Geometric Distribution: Inspections Until Defect')
plt.xlabel('Number of Inspections')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

Poisson Distribution:

The Poisson distribution models the count of events occurring in a fixed interval, where events are rare and independent.

Transportation Engineering: Number of vehicle arrivals or gaps in traffic per minute at an uncongested intersection, commonly used for traffic modeling.
Construction Management: Number of equipment failures per month on a site, aiding in maintenance scheduling within logistic models.
Environmental Engineering: Number of extreme weather events (e.g., floods) per year in a region, used in climate risk forecasting.

Problem: Vehicle arrivals at a busy intersection follow a Poisson distribution with an average rate of 5 vehicles per minute. Simulate arrivals for 60 minutes and calculate the probability of more than 7 arrivals in a minute.

```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
lam = 5
data = np.random.poisson(lam, 60)
prob_more_7 = 1 - poisson.cdf(7, lam)
print(f"Generated data (first 10 minutes): {data[:10]}")
print(f"Mean arrivals per minute: {np.mean(data):.2f}")
print(f"Probability of >7 arrivals: {prob_more_7:.4f}")

# Plot distribution (histogram with PMF overlay)
sns.histplot(data, bins=range(max(data)+2), stat='probability', kde=False)
x = np.arange(0, max(data)+1)
plt.plot(x, poisson.pmf(x, lam), 'ro-', label='PMF')
plt.title('Poisson Distribution: Vehicle Arrivals per Minute')
plt.xlabel('Number of Arrivals')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

Uniform Distribution:

The Uniform distribution models variables with equal probability across a range.

Transportation Engineering: Random arrival time of vehicles within a fixed time window (e.g., 0-60 seconds), assuming constant flow in simulations.
Construction Management: Uniform distribution of load across a beam in structural analysis, or random selection of inspection times during a shift.
Environmental Engineering: Wind direction in degrees (0-360), often modeled as uniform in initial assessments of dispersion models.

Problem: The time for a train to pass a level crossing is uniformly distributed between 30 and 60 seconds. Simulate 50 train passages and find the probability that a passage takes less than 45 seconds.

```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
a, b = 30, 60
data = np.random.uniform(a, b, 50)
prob_less_45 = (45 - a) / (b - a)
empirical_prob = np.mean(data < 45)
print(f"Generated data (first 10): {data[:10]}")
print(f"Theoretical P(<45): {prob_less_45:.2f}")
print(f"Empirical P(<45): {empirical_prob:.2f}")

# Plot distribution (histogram with PDF overlay)
sns.histplot(data, bins=10, stat='density', kde=False)
x = np.linspace(a, b, 100)
plt.plot(x, uniform.pdf(x, a, b-a), 'r-', label='PDF')
plt.title('Uniform Distribution: Train Passage Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Normal Distribution:

The Normal distribution models symmetric, bell-shaped data around a mean.

Transportation Engineering: Vehicle speeds on a highway, where most cluster around the average with symmetric deviations.
Construction Management: Measurement errors in material dimensions, such as concrete slab thickness, in quality control.
Environmental Engineering: Daily temperature variations in a region, used for forecasting trends and infrastructure design.

Problem: Vehicle speeds on a highway are normally distributed with mean 65 mph and std dev 5 mph. Simulate 200 speeds and find the probability that a speed exceeds 70 mph (speeding threshold).

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
mu, sigma = 65, 5
data = np.random.normal(mu, sigma, 200)
prob_exceed_70 = 1 - norm.cdf(70, mu, sigma)
empirical_prob = np.mean(data > 70)
print(f"Generated data (first 10): {data[:10]}")
print(f"Mean speed: {np.mean(data):.2f} mph")
print(f"Theoretical P(>70): {prob_exceed_70:.4f}")
print(f"Empirical P(>70): {empirical_prob:.4f}")

# Plot distribution (histogram with PDF overlay)
sns.histplot(data, bins=20, stat='density', kde=False)
x = np.linspace(min(data), max(data), 100)
plt.plot(x, norm.pdf(x, mu, sigma), 'r-', label='PDF')
plt.title('Normal Distribution: Vehicle Speeds')
plt.xlabel('Speed (mph)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

t Distribution:

The t distribution models means from small samples, similar to normal but with heavier tails.

Transportation Engineering: Estimated mean travel time from a small sample of 10 vehicles on a route, accounting for variability in traffic studies.
Construction Management: Mean project cost from a limited number of similar past projects, used when data is scarce.
Environmental Engineering: Mean pollutant concentration from a small set of water samples, applied in preliminary site assessments.

Problem: In transportation engineering, a small sample of 10 vehicles is tested for braking distance on a wet road to assess pavement friction. The braking distances (in meters) are assumed to follow a t-distribution due to the small sample size, with a mean of 35 meters and a scale parameter of 3 meters (reflecting variability). Simulate these distances and estimate the probability that a vehicle's braking distance exceeds 40 meters (a safety threshold for road design).

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
df = 9  # degrees of freedom (n-1 for 10 vehicles)
loc = 35  # mean braking distance in meters
scale = 3  # scale parameter
data = t.rvs(df, loc=loc, scale=scale, size=10)
prob_exceed_40 = 1 - t.cdf(40, df, loc=loc, scale=scale)
empirical_prob = np.mean(data > 40)
print(f"Generated braking distances: {data}")
print(f"Sample mean: {np.mean(data):.2f} meters")
print(f"Theoretical P(>40): {prob_exceed_40:.4f}")
print(f"Empirical P(>40): {empirical_prob:.4f}")

# Plot distribution (histogram with PDF overlay)
sns.histplot(data, bins=5, stat='density', kde=False)
x = np.linspace(min(data)-5, max(data)+5, 100)
plt.plot(x, t.pdf(x, df, loc=loc, scale=scale), 'r-', label='PDF')
plt.axvline(40, color='g', linestyle='--', label='Threshold (40m)')
plt.title('t Distribution: Braking Distances')
plt.xlabel('Braking Distance (meters)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Log Normal Distribution:

The Log Normal distribution models skewed, positive-valued data, such as times or costs.

Transportation Engineering: Distribution of trip durations or delays, where small times are common but long delays create right skew.
Construction Management: Schedule overruns or project completion times, as delays accumulate multiplicatively.
Environmental Engineering: Pollutant concentrations in air or water, which are positive and often log-skewed due to varying emission sources.

Problem: Repair times for traffic signals are log-normally distributed (skewed right) with mean log-time 1.5 (e^1.5 ~4.48 hours) and sigma 0.5. Simulate 100 repairs and find the median repair time.

```python
import numpy as np
from scipy.stats import lognorm
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
mu, sigma = 1.5, 0.5
data = np.random.lognormal(mu, sigma, 100)
median = np.exp(mu)
emp_median = np.median(data)
print(f"Generated data (first 10): {data[:10]}")
print(f"Theoretical median: {median:.2f} hours")
print(f"Empirical median: {emp_median:.2f} hours")

# Plot distribution (histogram with PDF overlay)
sns.histplot(data, bins=20, stat='density', kde=False)
x = np.linspace(0, max(data), 100)
plt.plot(x, lognorm.pdf(x, sigma, scale=np.exp(mu)), 'r-', label='PDF')
plt.title('Log Normal Distribution: Repair Times')
plt.xlabel('Time (hours)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Chi Square Distribution:

The Chi Square distribution is used for variance testing or goodness-of-fit.

Transportation Engineering: Testing if observed vehicle type frequencies (e.g., cars, trucks) match expected proportions in traffic composition studies.
Construction Management: Testing variance in material strength (e.g., concrete compressive strength) against a target value in quality assurance.
Environmental Engineering: Goodness-of-fit test for observed rainfall data against a theoretical model, or variance in sea level measurements.

Problem: To test variance in asphalt thickness measurements, 15 samples have variance 4 mm^2. Use chi-square to test if population variance exceeds 3 mm^2 (H0: sigma^2 <=3).

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
n = 15
df = n - 1
sample_var = 4  # mm^2
target_var = 3
chi_stat = df * sample_var / target_var
p_val = 1 - chi2.cdf(chi_stat, df)
# Generate sample data for illustration
data = np.random.normal(10, np.sqrt(sample_var), n)  # Assume mean 10 mm
print(f"Generated thickness data: {data}")
print(f"Sample variance: {np.var(data, ddof=1):.2f}")
print(f"Chi-square statistic: {chi_stat:.2f}")
print(f"P-value: {p_val:.4f} (if <0.05, reject H0)")

# Plot distribution (histogram of simulated chi-square values with PDF overlay)
sim_chi = np.random.chisquare(df, 1000)
sns.histplot(sim_chi, bins=30, stat='density', kde=False)
x = np.linspace(0, max(sim_chi), 100)
plt.plot(x, chi2.pdf(x, df), 'r-', label='PDF')
plt.axvline(chi_stat, color='g', label='Test Statistic')
plt.title('Chi-Square Distribution: Asphalt Thickness Variance')
plt.xlabel('Chi-Square Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Chi Square Distribution:

In transportation engineering, historical data from a urban highway indicates that the distribution of vehicle types during peak hours is: 60% cars, 25% SUVs, 10% trucks, and 5% motorcycles. To check if recent traffic patterns have changed, a sample of 400 vehicles is observed, with counts: 220 cars, 110 SUVs, 50 trucks, and 20 motorcycles. Use the Chi-Square goodness-of-fit test to determine if the observed sample matches the expected historical distribution (at alpha=0.05 significance level). Simulate the sample data based on slightly deviated proportions for realism.

```python
import numpy as np
from scipy.stats import chi2, chisquare
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)

# Expected proportions
expected_props = np.array([0.60, 0.25, 0.10, 0.05])
categories = ['Cars', 'SUVs', 'Trucks', 'Motorcycles']
n = 400
expected_counts = expected_props * n

# Generate observed counts (slightly deviated for realism)
observed_counts = np.random.multinomial(n, [0.55, 0.275, 0.125, 0.05])

# Perform Chi-Square goodness-of-fit test
chi_stat, p_val = chisquare(observed_counts, expected_counts)
df = len(categories) - 1
critical_val = chi2.ppf(0.95, df)
conclusion = "Reject H0 (distributions differ)" if p_val < 0.05 else "Fail to reject H0 (distributions match)"

print(f"Expected counts: {expected_counts}")
print(f"Observed counts: {observed_counts}")
print(f"Chi-square statistic: {chi_stat:.2f}")
print(f"P-value: {p_val:.4f}")
print(f"Critical value (alpha=0.05): {critical_val:.2f}")
print(f"Conclusion: {conclusion}")

# Plot observed vs expected (bar plot)
plt.figure(figsize=(10, 5))
width = 0.35
x = np.arange(len(categories))
plt.bar(x - width/2, expected_counts, width, label='Expected')
plt.bar(x + width/2, observed_counts, width, label='Observed')
plt.xticks(x, categories)
plt.title('Observed vs Expected Vehicle Types')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.legend()
plt.show()

# Plot Chi-Square distribution with test statistic
x = np.linspace(0, 20, 1000)
plt.plot(x, chi2.pdf(x, df), 'r-', label='Chi-Square PDF')
plt.axvline(chi_stat, color='g', linestyle='--', label='Test Statistic')
plt.axvline(critical_val, color='b', linestyle='--', label='Critical Value')
plt.title('Chi-Square Distribution (df=3)')
plt.xlabel('Chi-Square Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Exponential Distribution:

The Exponential distribution models time between events in a Poisson process.

Transportation Engineering: Time headways between consecutive vehicles in traffic flow, especially in low-density conditions.
Construction Management: Time between arrival of supply deliveries on site, used in logistic simulations.
Environmental Engineering: Time between extreme events like storms or wave surges, in marine or climate modeling.

Problem: The time between consecutive bus arrivals at a stop is exponentially distributed with mean 10 minutes. Simulate 100 inter-arrival times and find the probability that the next bus arrives within 5 minutes.

```python
import numpy as np
from scipy.stats import expon
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
mean_time = 10  # minutes
lam = 1 / mean_time
data = np.random.exponential(1/lam, 100)
prob_within_5 = 1 - np.exp(-lam * 5)
emp_prob = np.mean(data <= 5)
print(f"Generated data (first 10): {data[:10]}")
print(f"Mean inter-arrival: {np.mean(data):.2f} minutes")
print(f"Theoretical P(<=5): {prob_within_5:.4f}")
print(f"Empirical P(<=5): {emp_prob:.4f}")

# Plot distribution (histogram with PDF overlay)
sns.histplot(data, bins=20, stat='density', kde=False)
x = np.linspace(0, max(data), 100)
plt.plot(x, expon.pdf(x, scale=mean_time), 'r-', label='PDF')
plt.title('Exponential Distribution: Bus Inter-Arrival Times')
plt.xlabel('Time (minutes)')
plt.ylabel('Density')
plt.legend()
plt.show()
```
